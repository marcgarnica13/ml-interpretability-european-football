```python
#Import statements
import os
import pandas as pd
import numpy as np
```

```python
data_folder_path = os.path.join('/', 'home', 'marc', 'Development', 'dshs', 'gender_analysis', 'data')
data_file = 'ds.csv'
```

```python
os.path.join(data_folder_path, data_file)
```

```python
df = pd.read_csv(os.path.join(data_folder_path, data_file))
df.describe()
df.PlayerType.value_counts()
```

```python
df[df.matchPeriod == 'P'].head(10)
```

```python
columns_list = list(df.columns)
kpis_list = [l for l in columns_list if l not in ['matchPeriod', 'PlayerType', 'gender']]
len(kpis_list)
```

```python
df.shape
```

```python
data_dict = {}
for kpi in kpis_list:
    initial = {
        "Male_mean": np.round(df.loc[df['gender'] == 0,kpi].mean(), 1),
        "Male_std": np.round(df.loc[df['gender'] == 0,kpi].std(), 1),
        "Female_mean": np.round(df.loc[df['gender'] == 1,kpi].mean(), 1),
        "Female_std": np.round(df.loc[df['gender'] == 1,kpi].std(), 1)
    }
    data_dict[kpi.replace('_', ' ')] = {
        "Male": f"{initial['Male_mean']} ± {initial['Male_std']}",
        "Female": f"{initial['Female_mean']} ± {initial['Female_std']}"
    }
```

```python
data_dict
```

```python
stats = pd.DataFrame.from_dict(data=data_dict, orient='index')
stats
```

```python
stats = stats.loc[[kpi.replace('_', ' ') for kpi in kpis_list]].reset_index()
stats
```
